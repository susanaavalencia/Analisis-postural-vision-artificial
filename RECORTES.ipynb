```python
# ============================================================
# SUBIR VIDEO y corregir nombre si Colab pone (1)
# ============================================================
from google.colab import files
import subprocess
import json
import os
import cv2
import re

uploaded = files.upload()
input_video_raw = list(uploaded.keys())[0]
print("Archivo subido:", input_video_raw)

# Nombre sin extensión
base_original = re.sub(r" \(\d+\)$", "", os.path.splitext(input_video_raw)[0])
ext = os.path.splitext(input_video_raw)[1]
desired_name = base_original + ext

# Renombrar si Google agrega (1)
if input_video_raw != desired_name:
    if os.path.exists(desired_name):   # Si existe, se borra
        os.remove(desired_name)
    os.rename(input_video_raw, desired_name)
    input_video = desired_name
    print("Nombre corregido a:", desired_name)
else:
    input_video = input_video_raw

base_name = os.path.splitext(input_video)[0]


# ============================================================
# DETECTAR SI EL VIDEO ES HORIZONTAL
# ============================================================
def is_horizontal(video_path):
    cap = cv2.VideoCapture(video_path)
    w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    cap.release()
    return w > h


# ============================================================
# EXTRAER FRAGMENTOS SIN QUE SE GENEREN (1)
# ============================================================
def extract_fragment(input_video, start_sec, duration_sec, output_name):

    # Borrar el fragmento previo si existe
    if os.path.exists(output_name):
        os.remove(output_name)

    horizontal = is_horizontal(input_video)

    if horizontal:
        vf_filter = "transpose=1,scale=720:1280"
    else:
        vf_filter = "scale=720:1280"

    cmd = [
        "ffmpeg",
        "-y",
        "-ss", str(start_sec),
        "-i", input_video,
        "-t", str(duration_sec),
        "-vf", vf_filter,
        "-an",
        "-r", "30",
        "-b:v", "900k",
        "-c:v", "libx264",
        "-preset", "fast",
        output_name
    ]

    subprocess.run(cmd)


# ============================================================
# INTERVALOS A RECORTAR
# ============================================================
intervalos = [(1,7),(7,13),(13,18),(18,25),(25,32)]

# ============================================================
# 5️⃣ GENERAR FRAGMENTOS
# ============================================================
for i, (start, end) in enumerate(intervalos, start=1):
    duration = end - start
    output_name = f"{base_name}_Rep_{i}.mp4"

    extract_fragment(input_video, start, duration, output_name)
    print(f"Generado: {output_name}")


# ============================================================
# DESCARGAR FRAGMENTOS
# ============================================================
for i in range(1, len(intervalos) + 1):
    files.download(f"{base_name}_Rep_{i}.mp4")
```
